```python
#training_data=[{"class":i,"sentence":a} for i in ['basic','general','rule','training'] for a in basic_list]
training_data=[{"class":"basic","sentence":i} for i in basic_list]
training_data=training_data+[{"class":"general","sentence":i} for i in general_list]
training_data=training_data+[{"class":"rule","sentence":i} for i in rule_list]
training_data=training_data+[{"class":"training","sentence":i} for i in training_list]
#training_data=[{"class":i,"sentence":a} for i in df.columns for a in df[i]]
training_data
```

```python
corpus_words = {}
#class_words=[]
classes = list(set([a['class'] for a in training_data]))

class_words={}
for c in classes:
    class_words[c] = []
corpus_words
```

```python
for data in training_data:
    for word in word_tokenize(data['sentence']):
        if word is " ":
            continue
        if word not in corpus_words:
            corpus_words[word] = 1
        else:
            corpus_words[word] += 1  
        class_words[data['class']].extend([word])

print ("Corpus words and counts: %s" % corpus_words)
print ("Class words: %s" % class_words)
```

```python
sentence = "วันไหนได้หยุดสงกรานต์บ้างครับ"
def calculate_class_score(sentence, class_name):
    score = 0
    for word in word_tokenize(sentence):
        if word in class_words[class_name]:
            score += 1
    return score

for c in class_words.keys():
    print ("Class: %s  Score: %s" % (c, calculate_class_score(sentence, c)))
```

```python
def calculate_class_score_commonality(sentence, class_name):
    score = 0
    for word in word_tokenize(sentence):
        if word in class_words[class_name]:
            score += (1 / corpus_words[word])
    return score

for c in class_words.keys():
    print ("Class: %s  Score: %s" % (c, calculate_class_score_commonality(sentence, c)))
```

```python
def find_class(sentence):
    high_class = None
    high_score = 0
    for c in class_words.keys():
        score = calculate_class_score_commonality(sentence, c)
        if score > high_score:
            high_class = c
            high_score = score
    return high_class, high_score
find_class(sentence)
```
